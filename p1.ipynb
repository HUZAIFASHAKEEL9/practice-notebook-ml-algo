```python
import numpy as np
from sklearn.linear_model import LinearRegression, LogisticRegression
import matplotlib.pyplot as plt
```

```python
stamp_bought = np.array([2, 5, 7, 4]).reshape((-1,1))
amount_spent = np.array([1, 3, 4, 2])

model = LinearRegression()
model.fit(stamp_bought, amount_spent)

next_month = 10
predicted_amount = model.predict([[next_month]])

plt.scatter(stamp_bought, amount_spent, color='blue')
plt.plot(stamp_bought, model.predict(stamp_bought),color='red')
plt.xlabel("Stamp buy")
plt.ylabel("Amount $")
plt.grid(True)
plt.show()

print(f'If Ahmed buys {next_month} stamps so the prediction for amount is ${predicted_amount[0]:.2f}.')
```

```python
book_pages = np.array([50, 100, 150, 200, 250, 300, 350, 450, 500]).reshape((-1,1))
is_interested = np.array([0, 1, 1, 1, 1, 1, 0, 0, 0])

model = LogisticRegression()
model.fit(book_pages, is_interested)

book_of = 251

predicted_like = model.predict([[book_of]])

plt.scatter(book_pages, is_interested, color='blue')
plt.plot(book_pages, model.predict(book_pages),color='red')
#for reference line on plot
plt.axvline(x=predicted_amount, color='green', linestyle='--')
plt.axhline(y=0.5, color='grey', linestyle='--')
plt.xlabel("Book having pages")
plt.ylabel("Insteres or not?")
plt.grid(True)
plt.show()

print(f'If jenny read book of pages {book_of}, {'he likes' if predicted_like[0] == 1 else 'not like'}')
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
fruits_feature = np.array([[2, 7], [3, 6], [4, 8], [5, 7], [1, 3], [4, 2], [7, 6], [6, 6]]) # size and sweetness
fruit_like = np.array([1, 1, 1, 1, 0, 0, 0, 1])

model = LinearDiscriminantAnalysis()
model.fit(fruits_feature, fruit_like)

new_fruit = np.array([[2.5, 4]])
predict_like = model.predict(new_fruit)

plt.scatter(fruits_feature[:, 0], fruits_feature[:,1], c=fruit_like, cmap='viridis', marker='o')
plt.scatter(new_fruit[:,0], new_fruit[:,1], color='darkred', marker='x')
plt.xlabel("Fruit Features")
plt.ylabel("Like?")
plt.grid(True)
plt.show()

print(f'If Sarah has fruit of size{new_fruit[0,0]} of size {new_fruit[0, 1]} sarah will {'like' if predict_like == 0 else 'not like'}.')
```
